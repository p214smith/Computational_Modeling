## Background

### Rules

All cells (infected or not) will follow "Conway's Game of Life" rules. This means if a living cell as less than two or more than four living neighbors it will die. If a living cell has 2 or 3 living neighbors it will stay alive. If a dead cell has 3 neighbors it will come alive. When the simulation begins there will be an infector cell that will begin the spread of the virus. If a cell becomes infected, a check will be done to see if the cell survives the infection during that simulation loop. If the infected cell is marked for death it will die regardless of the rules for "Conways Game of Life." If a cell dies due to infection It will be added to the total of infected deaths. If the infected cell survives a designated amount of simulation loops it will be change state to uninfected and be added to a total count of survivors. Any uninfected cells that are adjacent to an infected cell have a possibility to be infected. Once the simulation is complete statistics for the deaths from infection, and survival rates will be displayed.

### Assumptions

I am making several assumptions about the spread of infection within the game. The largest assumption I am making is I am not accounting for immunity. While I understand immunity is a key to infectious disease spread, it is not a factor I believe I can account for without making serious design choice changes. This means that a cell that beats the infection has a possibility of becoming reinfected in the next step. Another assumption I am making is that if a cell dies while infected, but the death is not related to the infection, i.e. due purely to the rules of the game of life, this death will not be counted as a death from infection. I am choosing to track deaths this way to limit the amount of deaths that may be misrepresented. The only way that a cell can be marked as a survivor if the cell survives a number of steps without dying from infection or due to the rules of the game of life. This ensures that all infection statistics are gathered from cells that had the opportunity to survive a full length infection.

```python
class the_game_of_infected:
    def __init__(self,grid,stills,ships,penta,infectors,mortality,spread,length):
        self.grid = np.empty((grid,grid),dtype = box)
        self.infection_length = length
        i = 0
        j = 0
        while i < grid:
            j = 0
            while j < grid:
                self.grid[i][j] = box(size= vector(1,1,1),pos= vector(i - grid/2,j-grid/2,0), color = color.blue)
                self.grid[i][j].state = 0
                self.grid[i][j].next_state = 0
                self.grid[i][j].clear_infection = self.infection_length
                self.grid[i][j].infected = 0
                j +=1
            i+=1
        self.still = stills
        self.ship = ships
        self.pent = penta
        self.infector = infectors
        self.mortality = mortality
        self.n = grid
        self.spread = spread
        self.survived_infection = 0
        self.did_not_survive = 0
        
    def ships(self):
        start = random.randint(1,self.n -11)
        start1 = random.randint(1,self.n -11)
        self.grid[start][start1+10].state = 1
        self.grid[start+1][start1+10].state = 1
        self.grid[start+2][start1+10].state = 1
        self.grid[start+3][start1+10].state = 1
        self.grid[start+4][start1+10].state = 1
        self.grid[start+5][start1+10].state = 1
        self.grid[start-1][start1+9].state = 1
        self.grid[start+5][start1+9].state = 1
        self.grid[start+5][start1+8].state = 1
        self.grid[start-1][start1+7].state = 1
        self.grid[start+4][start1+7].state = 1
        self.grid[start+1][start1+6].state = 1
        self.grid[start+2][start1+6].state = 1
        self.grid[start][start1+10].color = color.white
        self.grid[start+1][start1+10].color = color.white
        self.grid[start+2][start1+10].color = color.white
        self.grid[start+3][start1+10].color = color.white
        self.grid[start+4][start1+10].color = color.white
        self.grid[start+5][start1+10].color = color.white
        self.grid[start-1][start1+9].color = color.white
        self.grid[start+5][start1+9].color = color.white
        self.grid[start+5][start1+8].color = color.white
        self.grid[start-1][start1+7].color = color.white
        self.grid[start+4][start1+7].color = color.white
        self.grid[start+1][start1+6].color = color.white
        self.grid[start+2][start1+6].color = color.white
        
    def stills(self):
        start = random.randint(1,self.n -2)
        start1 = random.randint(1,self.n -2)
        self.grid[start][start1].state = 1
        self.grid[start][start1+1].state = 1
        self.grid[start+1][start1].state = 1
        self.grid[start+1][start1+1].state = 1
        self.grid[start][start1].color = color.white
        self.grid[start][start1+1].color = color.white
        self.grid[start+1][start1].color = color.white
        self.grid[start+1][start1+1].color = color.white
        
    def penta(self):
        start = random.randint(1,self.n -5)
        start1 = random.randint(1,self.n -31)
        self.grid[start][start1+30].state = 1
        self.grid[start+1][start1+30].state = 1
        self.grid[start+2][start1+30].state = 1
        self.grid[start-1][start1+28].state = 1
        self.grid[start+3][start1+28].state = 1
        self.grid[start-1][start1+27].state = 1
        self.grid[start+3][start1+27].state = 1
        self.grid[start][start1+25].state = 1
        self.grid[start+1][start1+25].state = 1
        self.grid[start+2][start1+25].state = 1
        self.grid[start][start1+22].state = 1
        self.grid[start+1][start1+22].state = 1
        self.grid[start+2][start1+22].state = 1
        self.grid[start-1][start1+20].state = 1
        self.grid[start+3][start1+20].state = 1
        self.grid[start-1][start1+19].state = 1
        self.grid[start+3][start1+19].state = 1
        self.grid[start][start1+17].state = 1
        self.grid[start+1][start1+17].state = 1
        self.grid[start+2][start1+17].state = 1
        self.grid[start][start1+30].color = color.white
        self.grid[start+1][start1+30].color = color.white
        self.grid[start+2][start1+30].color = color.white
        self.grid[start-1][start1+28].color = color.white
        self.grid[start+3][start1+28].color = color.white
        self.grid[start-1][start1+27].color = color.white
        self.grid[start+3][start1+27].color = color.white
        self.grid[start][start1+25].color = color.white
        self.grid[start+1][start1+25].color = color.white
        self.grid[start+2][start1+25].color = color.white
        self.grid[start][start1+22].color = color.white
        self.grid[start+1][start1+22].color = color.white
        self.grid[start+2][start1+22].color = color.white
        self.grid[start-1][start1+20].color = color.white
        self.grid[start+3][start1+20].color = color.white
        self.grid[start-1][start1+19].color = color.white
        self.grid[start+3][start1+19].color = color.white
        self.grid[start][start1+17].color = color.white
        self.grid[start+1][start1+17].color = color.white
        self.grid[start+2][start1+17].color = color.white
    
    def infectors(self):
        start = random.randint(0,self.n -2)
        start1 = random.randint(0,self.n -2)
        self.grid[start][start1].state = 1
        self.grid[start][start1+1].state = 1
        self.grid[start+1][start1].state = 1
        self.grid[start+1][start1+1].state = 1
        self.grid[start][start1].infected = 1
        self.grid[start][start1+1].infected = 1
        self.grid[start+1][start1].infected = 1
        self.grid[start+1][start1+1].infected = 1
        self.grid[start][start1].color = color.green
        self.grid[start][start1+1].color = color.green
        self.grid[start+1][start1].color = color.green
        self.grid[start+1][start1+1].color = color.green
        self.grid[start][start1].clear_infection = 30
        self.grid[start][start1+1].clear_infection = 30
        self.grid[start+1][start1].clear_infection = 30
        self.grid[start+1][start1+1].clear_infection = 30
        
    def update_grid(self):
        i = 0
        
        while i < self.n:
            j = 0
            while j < self.n:
                left = i -1
                right = i + 1
                up = j + 1
                down = j - 1
                if left == -1:
                    left = self.n-1
                if right > self.n-1:
                    right = 0
                if down == -1:
                    down = self.n-1
                if up > self.n-1:
                    up = 0
                neighbor = self.grid[left][up].state + self.grid[left][j].state + self.grid[left][down].state + self.grid[i][up].state + self.grid[i][down].state + self.grid[right][up].state + self.grid[right][j].state + self.grid[right][down].state
                if self.grid[i][j].state == 0 and neighbor == 3:
                    self.grid[i][j].next_state = 1
                elif self.grid[i][j].state == 1 and (neighbor == 2 or neighbor == 3):
                    self.grid[i][j].next_state = 1
                else:
                    self.grid[i][j].next_state = 0
                if self.grid[i][j].infected == 1 :
                    if (self.grid[left][up].state == 1 and self.grid[left][up].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[left][up].infected = 1
                    if (self.grid[left][j].state == 1 and self.grid[left][j].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[left][j].infected = 1
                    if (self.grid[left][down].state == 1 and self.grid[left][down].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[left][down].infected = 1
                    if (self.grid[i][up].state == 1 and self.grid[i][up].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[i][up].infected = 1
                    if (self.grid[i][j].state == 1 and self.grid[i][j].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[i][j].infected = 1
                    if (self.grid[i][down].state == 1 and self.grid[i][down].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[i][down].infected = 1
                    if (self.grid[right][up].state == 1 and self.grid[right][up].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[right][up].infected = 1
                    if (self.grid[right][j].state == 1 and self.grid[right][j].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[right][j].infected = 1
                    if (self.grid[right][down].state == 1 and self.grid[right][down].infected == 0):
                        if random.randint(1,self.spread) != 1:
                            self.grid[right][down].infected = 1
                    if random.randint(1,self.mortality) == 1 :
                        self.did_not_survive += 1
                        self.grid[i][j].infected = 0
                        self.grid[i][j].state = 0
                        self.grid[i][j].next_state = 0
                        self.grid[i][j].clear_infection = self.infection_length
                    else:
                        self.grid[i][j].clear_infection -= 1
                        if self.grid[i][j].clear_infection == 0:
                            self.survived_infection += 1
                            self.grid[i][j].infected = 0
                            self.grid[i][j].clear_infection = self.infection_length
                    
                j +=1
            i +=1
        i = 0
        j = 0
        while i < self.n:
            j = 0
            while j < self.n:
                self.grid[i][j].state = self.grid[i][j].next_state
                if self.grid[i][j].state == 1 and self.grid[i][j].infected == 1:
                    self.grid[i][j].color = color.green
                elif self.grid[i][j].state == 1:
                    self.grid[i][j].color = color.white
                else:
                    self.grid[i][j].color = color.blue
                j +=1
            i+=1
```

## Evaluation

Due to the simplicity of my model there are numerous limitations to it. One of the first limitations I noticed was because there is no immunity reaction to the model still lifes that are not being disturbed were either continually being reinfected or dying off. Another limitation of my system is I'm not tracking infections of cells that die off due to the rules of the game of life. I felt that this design choice presented me with a more accurate account of the spread. By only tracking the spread cells that either died exclusively to the infection or survived the infection completely I was better able to calculate the mortality rate. I believe this can be shown with my simulation that I ran with a calculated flu rate. By being able to calculate the correct mortality rate to within 0.003% accuracy I was able to see how the infection lasted after extended runs. There were several other Interesting observations I was able to make while watching the simulation run. The grids with the higher rate of morality had a much lower rate of overall infection and the field of living cells died off much sooner. The grids with lower rates of mortality had much higher rates of overall infection given all other parameters were the same. There were more living cells at the end of the simulation and the infection tended to last the entire simulation. In the simulations where the mortality rate was much higher the infection tended to die off completely much earlier in the simulation and rarely made it to the end.

I think this simulation would be best suited for someone who was trying to understand how to get an early idea of how to create a more detailed model. It has the basis for a simple yet effective way to see how an infection can spread. You can use this to gather ideas about spread and some effective ways to model it.  When I was initially designing the model I wanted to have a roaming infector that would roam the grid constantly spreading the infection. I decided that would overcomplicate the system and make the simulation loop run longer than necessary. I decided to use numerous still lifes to intitialize the infection instead, which created a cleaner simulation loop while still spreading the infection throughout the grid. Keeping the simulation loop running relatively quickly was one of the biggest challenges I needed to overcome. I felt that I needed to iterate over the entire grid because of the density of the population and the need to spread the infection properly. Because of this I needed to find other places to cut the simulation time so I decided to simplify some of the initial ideas for the spread of the infection while still delivering a competent overall simulation. Given the course concepts I do feel like there could be a way to do an infection simulation based on discrete time as opposed to a step based simulation. 

## Conclusion

This project turned out a little different than I expected. When initially thought of this project I was thinking of going about how to approach the infection very differently. With a little trial and error I was able to come up with a system that I feel represented what I wanted to accomplish in the end. The first feature I would look to add would be an immunity state.
